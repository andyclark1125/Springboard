**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
#store the first row as the keys, merge back as a column later
keys = properties.iloc[[0]]
print(keys)
print(type(keys))
#remove the first row
properties.drop([0], inplace=True)
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#rename to date column
properties.rename(columns = {'Unnamed: 0' : 'date'}, inplace=True)
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#remove NaN columns before melting
nan_cols = ['Unnamed: 34', 'Unnamed: 37', 'Unnamed: 47']
properties.drop(columns= nan_cols, inplace=True)
properties.columns
```

```python
#melt properties from wide to tall format
properties = properties.melt(id_vars = ['date'], var_name= 'Boroughs', value_name='HousingPrice')
print(properties.shape)
print(properties.head())
```

```python
#remove date col from keys
keys.drop(keys.columns[[0]], axis=1, inplace=True)
keys
```

```python
#remove columns with all NaN values from keys
nan_cols = ['Unnamed: 34', 'Unnamed: 37', 'Unnamed: 47']
keys.drop(columns= nan_cols, inplace=True)
keys
```

```python
#melt keys into keys_tall
keys_tall = keys.melt(var_name=['Boroughs'], value_name='Key')
keys_tall
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
properties.info()
```

```python
#convert housing price column from object to float
properties['HousingPrice'] = pd.to_numeric(properties['HousingPrice'])

properties.info()
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
#merge df_tall and keys_tall into properties
properties = properties.merge(keys_tall, on='Boroughs', how='left')
print(properties.head())
properties.isna().any()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
properties.set_index('date', inplace=True)
properties.groupby('Boroughs')['HousingPrice'].plot(legend=True)
```

```python
city_of_london = properties[properties['Boroughs']=='City of London']
city_of_london.plot(y='HousingPrice')
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column.

```python
properties['year'] = properties.index.year
print(properties.head())
print(properties.tail())

#create a new panda data frame with average price by year
PriceByYear = properties.groupby(['Boroughs','year'])[['HousingPrice']].mean()
```

```python
type(PriceByYear)
print(PriceByYear.head())
print(PriceByYear.tail())
print(PriceByYear.info())
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
#reset index on PriceByYear
PriceByYear = PriceByYear.reset_index()
PriceByYear.columns

```

```python
PriceByYear['Boroughs'] = PriceByYear['Boroughs'].astype(str)
print(PriceByYear.info())
print(PriceByYear.head())
```

```python
#query the PriceByYear for the specific years, then calculate the ratio in descending order
StartEndPrice = PriceByYear.query('year == 1998 or year == 2018')
StartEndPrice_wide = StartEndPrice.pivot(index='Boroughs', columns='year', values='HousingPrice')
StartEndPrice_wide['ratio'] = StartEndPrice_wide[2018] / StartEndPrice_wide[1998]
StartEndPrice_wide.sort_values(by='ratio', ascending = False, inplace=True)
StartEndPrice_wide.head()
```

```python
StartEndPrice_wide.reset_index(inplace=True)
StartEndPrice_wide.iloc[0:10,[0,3]].plot(kind='bar',x='Boroughs',\
y='ratio', ylabel='Price Ratio', title='Highest 10 London Boroughs Housing Price Ratio 2018 / 1998',\
grid=True)
```
